```python
gluu = 'https://auth-jsonb.alpha.vaultit.org'
qvarn = 'http://localhost:9001'
planb = 'http://localhost:9002'
jsonb = 'http://localhost:9003'

names = {
    qvarn: 'qvarn',
    planb: 'planb',
    jsonb: 'jsonb',
}

backends = qvarn, jsonb, planb
```

```python
import collections
import configparser
import faker
import hashlib
import logging as log
import os
import random
import requests
import statistics
import sys
import time
import tqdm
import pandas as pd
import seaborn as sns
from IPython.display import display

log.basicConfig(filename='/tmp/performance-testing.log', level=log.DEBUG)
log.getLogger('urllib3').setLevel(log.ERROR)

sns.set(style="whitegrid", font_scale=2, rc={
    'figure.figsize': (16, 8),
})

gluu_config = requests.get(f'{gluu}/oxauth/.well-known/openid-configuration').json()

credentials_config = configparser.ConfigParser()
credentials_config.read(os.path.expanduser('~/.config/qvarn/createtoken.conf'))
credentials = credentials_config[gluu]

scopes = [
    scope
    for scope in gluu_config['scopes_supported']
    if scope.startswith(('uapi_contracts', 'uapi_orgs'))
]
data = {
    'grant_type': 'client_credentials',
    'scope': ' '.join(scopes),
}
auth = (credentials['client_id'], credentials['client_secret'])
resp = requests.post(f'{gluu}/oxauth/seam/resource/restv1/oxauth/token', data, auth=auth)
token = resp.json()['access_token']

session = requests.Session()
session.headers['authorization'] = f'Bearer {token}'

fake = faker.Faker()
```

```python
def measure(n, backends, test, *args, **kwargs):
    failed = {backend: 0 for backend in backends}
    times = collections.defaultdict(list)
    runs = {}
    for backend in backends:
        run = runs[backend] = test(n, backend, *args, **kwargs)
        name = names[backend]
        for item in tqdm.tqdm(run, name, total=run.total):
            start = time.monotonic()
            try:
                resp = run(item)
            except KeyboardInterrupt:
                raise
            except:
                log.exception('error')
                ok = False
                resp = None
            else:
                ok = resp.ok
            delta = time.monotonic() - start
            if ok:
                times[backend].append(delta)
            elif resp:
                log.error(resp.text)
                failed[backend] += 1
                
    if len([1 for x in backends if times[x]]) == 0:
        return
    
    frame = pd.DataFrame(collections.OrderedDict([
        (('iterations', 'total'), [runs[x].total for x in backends if times[x]]),
        (('iterations', 'failed'), [failed[x] for x in backends if times[x]]),
        (('time in seconds', 'total'), [sum(times[x]) for x in backends if times[x]]),
        (('time in seconds', 'mean'), [statistics.mean(times[x]) for x in backends if times[x]]),
        (('time in seconds', 'median'), [statistics.median(times[x]) for x in backends if times[x]]),
        (('time in seconds', 'stdev'), [statistics.stdev(times[x]) for x in backends if times[x]]),
        (('', 'number of resources'), [len(session.get(f'{x}/orgs').json()['resources']) for x in backends if times[x]]),
    ]), index=[names[x] for x in backends if times[x]]).sort_values(('time in seconds', 'total'))
    display(frame)
    
    size = max(len(times[x]) for x in backends if times[x])
    if size > 1:
        frame = pd.DataFrame(collections.OrderedDict([
            (names[x], times[x] + [None for i in range(size - len(times[x]))])
            for x in backends if times[x]
        ]))
        display(sns.violinplot(data=frame, orient='h'))
```

```python
def get_new_id(resource_type, random_field=None):
    type_field = hashlib.sha512(resource_type.encode()).hexdigest()[:4]
    random_field = random_field or os.urandom(16).hex()
    checksum_field = hashlib.sha512((type_field + random_field).encode()).hexdigest()[:8]
    return '{}-{}-{}'.format(type_field, random_field, checksum_field)

get_new_id('org')
```

```python
class Test:
    def __init__(self, n, backend, *args, **kwargs):
        self.backend = backend
        self.total = n
        self.prepare(*args, **kwargs)
        
    def prepare(self):
        pass
        
    def __iter__(self):
        for i in range(self.total):
            yield
            
    def __call__(self, item):
        raise NotImplementedError()
```

```python
class DeleteResources(Test):
    
    def prepare(self, resource_type):
        self.resource_type = resource_type
        self.resp = session.get(f'{self.backend}/{resource_type}').json()
        self.total = len(self.resp['resources'])
        
    def __iter__(self):
        for resource in self.resp['resources']:
            yield resource
            
    def __call__(self, resource):
        return session.delete(f'{self.backend}/{self.resource_type}/{resource["id"]}')
        
measure(None, backends, DeleteResources, 'orgs')
```

```python
class CreateOrgs(Test):
    
    def prepare(self):
        self.country = 'FI'
        
    def __call__(self, item):
        return session.post(f'{self.backend}/orgs', json={
            'type': 'org',
            'country': self.country,
            'names': (
                [fake.company(), fake.catch_phrase()]
                if random.randint(0, 50) == 42 else
                [fake.company()]
            ),
            'gov_org_ids': [
                {
                    'country': self.country,
                    'org_id_type': 'registration_number',
                    'gov_org_id': fake.ssn(),
                }
            ],
            'contacts': [
                {
                    'contact_type': 'phone',
                    'phone_number': fake.phone_number(),
                    'full_address': fake.address(),
                    'contact_timestamp': str(time.time()),
                    'country': fake.country_code(),
                    'contact_roles': [],
                    'address_lines': fake.address().splitlines(),
                    'post_code': None,
                    'post_area': None,
                    'einvoice_operator': None,
                    'contact_source': 'qvarn-genarator',
                    'einvoice_address': None,
                    'email_address': None,
                }
                for i in range(random.randint(1, 5))
            ],
        })
    
measure(100, backends, CreateOrgs)
```

```python
class GetAll(Test):
    def __call__(self, item):
        return session.get(f'{self.backend}/orgs')
    
measure(1000, backends, GetAll)
```

```python
class GetOne(Test):
    def prepare(self):
        self.resources = [
            resource['id'] for resource in session.get(f'{self.backend}/orgs').json()['resources']
        ]
    
    def __call__(self, item):
        return session.get(f'{self.backend}/orgs/{random.choice(self.resources)}')
    
measure(1000, backends, GetOne)
```

```python
class SearchExact(Test):
    def prepare(self):
        self.gov_org_ids = [
            resource['gov_org_ids'][0]['gov_org_id']
            for resource in session.get(f'{self.backend}/orgs/search/show/gov_org_ids').json()['resources']
        ]
    
    def __call__(self, item):
        return session.get(f'{self.backend}/orgs/search/exact/gov_org_id/{random.choice(self.gov_org_ids)}')
    
measure(100, backends, SearchExact)
```

```python
class SearchStartswith(Test):
    def prepare(self):
        self.gov_org_ids = [
            resource['gov_org_ids'][0]['gov_org_id']
            for resource in session.get(f'{self.backend}/orgs/search/show/gov_org_ids').json()['resources']
        ]
    
    def __call__(self, item):
        return session.get(f'{self.backend}/orgs/search/startswith/gov_org_id/{random.choice(self.gov_org_ids)}')
    
measure(100, backends, SearchStartswith)
```

```python
class SearchLimit(Test):
    def __call__(self, item):
        return session.get(f'{self.backend}/orgs/search/sort/id/limit/1')
    
measure(100, backends, SearchLimit)
```

```python
class SearchShow(Test):
    def __call__(self, item):
        return session.get(f'{self.backend}/orgs/search/show/names/sort/id/limit/100')
    
measure(100, backends, SearchShow)
```

```python
class SearchShowAll(Test):
    def __call__(self, item):
        return session.get(f'{self.backend}/orgs/search/show_all')
    
measure(100, backends, SearchShowAll)
```
